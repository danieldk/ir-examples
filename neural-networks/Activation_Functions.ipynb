```python
import math
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_function(f,w,b):
    x = np.arange(-10.0, 10.0, 0.1)
    plt.plot(x, f(x, w, b))
    plt.show()
    
def plot_double_function(f,w1,b1,w2,b2):
    x = np.arange(-10.0, 10.0, 0.1)
    plt.plot(x, f(x, w1, b1) + f(x, w2, b2) - 1.0)
    plt.show()
```

```python
def sigmoid(x, w, b): return 1. / (1 + np.exp(-(w*x+b)))
```

```python
interact(lambda w, b: plot_function(sigmoid, w, b), w=(-10.,10.,0.1), b=(-10.,10,0.1))
```

```python
interact(lambda w1, b1, w2, b2: plot_double_function(sigmoid, w1, b1, w2, b2), \
         w1=(-10.,10.,0.1),b1=(-10.,10,0.1),w2=(-10.,10.,0.1),b2=(-10.,10,0.1))
```

```python
def relu(x, w, b):
    return np.fmax(w*x+b, 0)
```

```python
interact(lambda w, b: plot_function(relu, w, b), w=(-10.,10.,0.1), b=(-10.,10,0.1))
```

```python
interact(lambda w1, b1, w2, b2: plot_double_function(relu, w1, b1, w2, b2), \
         w1=(-10.,10.,0.1),b1=(-10.,10,0.1),w2=(-10.,10.,0.1),b2=(-10.,10,0.1))
```
