```python
%matplotlib inline

import math
from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt
import numpy as np
from boundary_simple import *
```

```python
X = np.array([[0,0],[1,0],[0,1],[1,1]])
Y = np.array([0, 1, 1, 0])
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.Spectral)
```

```python
def sigmoid(x, w, b):
    return 1. / (1 + np.exp(-(np.dot(w, x)+b)))
```

```python
def plot_boundary(w1,w2,b,hw1_1,hw1_2,hb1,hw2_1,hw2_2,hb2):
    def f(x):
        return sigmoid(w1 * sigmoid(x, np.array([hw1_1, hw1_2]), hb1) + \
            w2 * sigmoid(x, np.array([hw2_1, hw2_2]), hb2) + b, 1, 0)
    plot_decision_boundary(f, X, Y)

def float_slider(v):
    return FloatSlider(min=-2, max=2, step=0.1, value=v)
    
interact(plot_boundary,
         w1=float_slider(0.3),
         w2=float_slider(-0.7),
         b=float_slider(0.5),
         hw1_1=float_slider(1.9),
         hw1_2=float_slider(-1.0),
         hb1=float_slider(-1.0),
         hw2_1=float_slider(1.6),
         hw2_2=float_slider(1.4),
         hb2=float_slider(1.0))
```

```python
def quadratic(x, w, b): return (np.dot(w, x)+b) ** 2

def plot_quadratic_boundary(w,b,hw1,hw2,hb):
    def f(x):
        return sigmoid(w * quadratic(x, np.array([hw1, hw2]), hb) + b, 1, 0)
    plot_decision_boundary(f, X, Y)

interact(plot_quadratic_boundary,
         w=float_slider(1.3),
         b=float_slider(-0.3),
         hw1=float_slider(-1.0),
         hw2=float_slider(1.1),
         hb=float_slider(-0.1))
```
